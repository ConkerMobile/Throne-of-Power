```python
def Battle(Player1, Player2):
    Player1.battle_health = Player1.health
    Player2.battle_health = Player2.health
    import tkinter as tk
    root = tk.Tk()
    frame = tk.Frame(root)
    frame.pack()
    canvas = tk.Canvas(root, width=1000, height=300)
    canvas.pack()
    canvas.pack()
    current_player = 1
    while True:
        
        status = canvas.create_text(500,100, text=f"""
             {Player1.name}                                                       {Player2.name}
             Health: {Player1.battle_health}                                                   Health: {Player2.battle_health}
             Total Damage: {Player1.max_damage}                                         Total Damage: {Player2.max_damage}
               """,
            tag = "test")
        
        
        def attack(move):
            nonlocal current_player
            print("THIS IS THE MOVE: " + move)
            if (current_player == 1):
                Player2.battle_health-= Player1.damage[move]
                current_player = 2
            else:
                Player1.battle_health-= Player2.damage[move]
                current_player = 1
            canvas.delete("test")
            status = canvas.create_text(500,100, text=f"""
             {Player1.name}                                                       {Player2.name}
             Health: {Player1.battle_health}                                                   Health: {Player2.battle_health}
             Total Damage: {Player1.max_damage}                                         Total Damage: {Player2.max_damage}
               """,
            tag = "test")
            return current_player
        
        for move in range(len(Player1.moves)):
            if (current_player == 1):
                print(move)
                move_button = tk.Button(frame,
                               text=f"{Player1.moves[move]}: {Player1.damage[Player1.moves[move]]}",
                               fg = 'green',
                               command = lambda: attack(Player1.moves[move]))
                
            else:
                move_button = tk.Button(frame,
                               text=f"{Player1.moves[move]}: {Player1.damage[Player1.moves[move]]}",
                               fg = 'black')
            move_button.pack(side=tk.LEFT)
        for move in range(len(Player2.moves)):
            if (current_player == 2):
                move_button = tk.Button(frame,
                               text=f"{Player2.moves[move]}: {Player2.damage[Player2.moves[move]]}",
                               fg = 'green',
                               command = lambda: attack(Player2.damage[Player2.moves[move]]))
            else:
                move_button = tk.Button(frame,
                               text=f"{Player2.moves[move]}: {Player2.damage[Player2.moves[move]]}",
                               fg = 'black')
            move_button.pack(side=tk.RIGHT)
        root.mainloop()
        

     
    # current_player = 1
    # while (True):
    #     clear_output(wait=True)
#          print(f"""
#          {Player1.name}                                                       {Player2.name}
#         {Player1.damage}           {Player2.damage} 
#            Health: {Player1.battle_health}                                                  Health: {Player2.battle_health}
#            Total Damage: {Player1.max_damage}                                       Total Damage: {Player2.max_damage}
#            """)
    #     if (Player1.battle_health <= 0):
    #         print(f"{Player2.name} WINS!!!!")
    #         Player2.wins+=1
    #         break
    #     if (Player2.battle_health <= 0):
    #         print(f"{Player1.name} WINS!!!!")
    #         Player1.wins+=1
    #         break
    #     if (current_player == 1):
    #         input_move = str(input(f"{Player1.name}| Which move do you choose: "))
    #         if (input_move in Player1.moves):
    #             Player2.battle_health-= Player1.damage[input_move]
    #             current_player = 2
    #             continue
    #     elif (current_player == 2):
    #         input_move = str(input(f"{Player2.name}| Which move do you choose: "))
    #         if (input_move in Player2.moves):
    #             Player1.battle_health-= Player2.damage[input_move]
    #             current_player = 1
    #             continue
```
